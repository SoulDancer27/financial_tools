```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
l = 100
```

```python
eta = np.random.normal(0, 1, l)
eps = np.random.normal(0, 5, l)
```

```python
plt.plot(eta)
plt.plot(eps)
```

```python
mu = np.zeros(l)
y = np.zeros(l)
```

```python
for i in range(1, l):
  mu[i] = mu[i-1] + eta[i-1]
  y[i] = mu[i] + eps[i]
```

```python
plt.plot(mu)
plt.plot(y)
```

```python
# Kalman gain
mu_est = np.zeros(100)
k = 0.15
```

```python
for i in range(1, l):
  mu_est[i] = (1-k) * mu_est[i-1] + k * y[i]
```

```python
plt.plot(mu_est)
plt.plot(mu)
```
