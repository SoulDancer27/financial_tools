```python
import pandas as pd
from datetime import datetime
```

```python
data = pd.read_csv("./data/IMOEX_140101_240124.csv", sep=";")
data.head()
```

```python
data.rename(columns={"<DATE>" : "date", "<TIME>": "time", 
                     "<OPEN>": "open", "<HIGH>": "high", 
                     "<LOW>": "low", "<CLOSE>": "close", 
                     "<VOL>": "vol"}, inplace=True)
```

```python
data.head()
```

```python
data.drop(["time"], axis=1, inplace=True)
```

```python
data = data[data["vol"] >0]
data.head()
```

```python
data["date"] = data.apply(lambda row: str(int(row["date"])), axis=1)
display(data)
```

```python
data["date"] = data.apply(lambda row: datetime.strptime(row["date"], "%y%m%d"), axis=1)
display(data)
```

```python
data.to_csv("./cleaned/imoex.csv")
```
